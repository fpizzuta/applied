# Build a nearest neighbor model for movie recommendation

We now create a nearest-neighbors model and apply it to movies data to find the movies that are closer to the Matrix 

```python
knn_model = graphlab.nearest_neighbors.create(movie_imdb,features=['director_name','genres','imdb_score','movie_facebook_likes'],label='movie_title')
```
